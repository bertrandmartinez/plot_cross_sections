```python
# Imports from python packages
import sys
import numpy as np
import matplotlib.pyplot as plt

# Imports from my packages (see the folder called utils)
sys.path.append('../utils/')
from bremsstrahlung import tot_cs_sp, tot_rcs_quad
from screening import Fermi_length

# Style of the plot
plt.style.use('../utils/minimal.style')
```

```python
# Physical parameters
Z = 47
Lf = Fermi_length(Z)
```

```python
# discretization of g1 axis
N = 100
axis_g1m1 = np.logspace(-3., 4., N)
axis_g1 = 1. + axis_g1m1
```

```python
# Compute the total cross-section
sigma = np.array([tot_cs_sp(Z, g1, Lf) for g1 in axis_g1])
phi = np.array([tot_rcs_quad(Z, g1, Lf) for g1 in axis_g1])
```

```python
# Figure
fig, axs = plt.subplots(1, 1, figsize=(6,4))
axs.plot(axis_g1m1, phi, label=r'$ \phi$')
axs.set_xscale('log')
axs.set_xlabel(r"$ \gamma_1 - 1 $")
axs.set_xlim([1.e-3, 1.e4])
axs.set_yscale('log')
axs.set_ylabel(r"$ \phi \, \rm (m^2) $")
axs.set_ylim([1.e-28, 1.e-22])
#axs.set_ylim([-0.1, 1.5])
plt.tight_layout()
#fig.savefig('../figures/bremsstrahlung_total_cross_section', bbox_inches='tight')

plt.show()
```

```python
# Figure
fig, axs = plt.subplots(1, 1, figsize=(6,4))
axs.plot(axis_g1m1, 0.5*sigma, label=r'$ \sigma$')
axs.set_xscale('log')
axs.set_xlabel(r"$ \gamma_1 - 1 $")
axs.set_xlim([1.e-3, 1.e4])
axs.set_yscale('log')
axs.set_ylabel(r"$ \sigma \, \rm (m^2) $")
axs.set_ylim([1.e-26, 1.e-22])
#axs.set_ylim([-0.1, 1.5])
plt.tight_layout()
#fig.savefig('../figures/bremsstrahlung_total_cross_section', bbox_inches='tight')

plt.show()
```
