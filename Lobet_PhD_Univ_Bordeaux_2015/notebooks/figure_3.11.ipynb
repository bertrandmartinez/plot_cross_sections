```python
# Imports from python packages
import sys
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import N_A

# Imports from my packages (see the folder called utils)
sys.path.append('../../utils/')
from bremsstrahlung import rad_cs_sp, rad_cs_dp
from screening import Fermi_length
from screening import fit_More_zstar

# Style of the plot
plt.style.use('../../utils/minimal.style')
```

```python
# Physical parameters
Z = 47
Lf = Fermi_length(Z)
```

```python
# discretization of g1 axis
N = 100
axis_g1m1 = np.logspace(-3., 4., N)
axis_g1 = 1. + axis_g1m1
```

```python
# Provide details on the plasma density
A = 63.546
rho = 8.96 # g/cm^3
ni = 1.e6 * N_A * rho / A # /m^3

# Provide details on the plasma temperature
Tplasma = [0., 1.e-2, 0.1, 1., 100.]

# Deduce the ionization degree
Zstar = np.array([ fit_More_zstar(Z, A, rho, T) for T in Tplasma ])
M = len(Tplasma)
```

```python
# Compute the radiative cross-section
phi_sp = np.array([rad_cs_sp(Z, g1, Lf) for g1 in axis_g1])
phi_dp = np.zeros([M,N])
for i in range(M):
    phi_dp[i] = np.array([rad_cs_dp(Z, g1, Lf, Tplasma[i], ni, Zstar[i]) for g1 in axis_g1])
```

```python
# Non-relativistic cross-section with Elwert correction
# We introduce on purpose the wrong term 1/beta_1 ** 2
p1 = np.sqrt(axis_g1 ** 2 - 1.)
b1 = np.sqrt(1.-(1. / axis_g1 ** 2))

phi_sp_error = phi_sp.copy()
phi_sp_error[axis_g1<2] = phi_sp_error[axis_g1<2] * (p1[axis_g1<2] / b1[axis_g1<2]) ** 2

phi_dp_error = phi_dp.copy()
for i in range(M):cc=['b','r','orange','purple', 'brown']
labels=[r"$0$", r"$0.01$", r"$0.1$", r"$1$", r"$100$"]

# Figure
fig, axs = plt.subplots(1, 1, figsize=(6,4))
axs.plot(axis_g1m1, phi_sp_error, c=cc[0], label=r'$ Z^* = 0$')
#axs.plot(axis_g1m1, phi_sp, c=cc[0], ls='--')
for i in range(1,M):
    axs.plot(axis_g1m1, phi_dp_error[i,:], c=cc[i], label=r"$T=$" + labels[i]+r"$ \, \rm keV$")
#    axs.plot(axis_g1m1, phi_dp[i,:], c=cc[i], ls ='--')
axs.set_xscale('log')
axs.set_xlabel(r"$ \gamma_1 - 1 $")
axs.set_xlim([1.e-3, 1.e4])
axs.set_yscale('log')
axs.set_ylabel(r"$ \phi \, \rm (m^2) $")
axs.set_ylim([1.e-28, 1.e-22])
#axs.set_ylim([-0.1, 1.5])
plt.legend(loc='best', ncol=1)
plt.tight_layout()
fig.savefig('../figures/figure_3_11', bbox_inches='tight')

plt.show()
    phi_dp_error[i,axis_g1<2] = phi_dp_error[i,axis_g1<2] * (p1[axis_g1<2] / b1[axis_g1<2]) ** 2
```
